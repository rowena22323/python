```python
import numpy as np
import pandas as pd
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
import platform
```

```python
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
df=pd.read_csv("../master/restaurante.csv", encoding='cp949')
df.head()
```

```python
service=df['업소명'].groupby(df['업태']).count()
srv=pd.DataFrame(service)
plt.plot(srv)
plt.show()
```

```python
filter1=df['시군구']=='거창군' 
filter2=df['시군구']=='함양'
filter3 = filter1|filter2
df3 = df[filter3]
df3.head()
```

```python
df2=df3.groupby(by='주메뉴').size().reset_index(name='건수')
df4=df2.sort_values(by='건수',ascending=False).tail(4)
plt.plot(df4['주메뉴'],df4['건수'])
plt.show()
```

```python
df6 = df[df['주메뉴'].str.match('낙지')]
df6.head()
```

```python
df7=df6.groupby(by='주메뉴').size().reset_index(name='건수')
df8=df7.sort_values(by='건수',ascending=False).head(3)
plt.plot(df8['주메뉴'],df8['건수'])
plt.show()
```
