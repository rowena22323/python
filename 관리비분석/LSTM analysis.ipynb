```python
data = publicBymonth.loc[:,['index','월별합계']]
data
```

```python
data.plot()
```

```python
train = data.iloc[0:70,:]
test = data.iloc[71:,:]
ax = train.plot()
test.plot(ax=ax)
plt.legend(['train', 'test'])
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()

train_sc = sc.fit_transform(train)
test_sc = sc.transform(test)

train_sc
```

```python
train_sc_df = pd.DataFrame(train_sc, columns=['월','월별합계'], index=train.index)
test_sc_df = pd.DataFrame(test_sc, columns=['월','월별합계'], index=test.index)
train_sc_df.head()
```

```python
for s in range(1, 13):
    train_sc_df['shift_{}'.format(s)] = train_sc_df['월별합계'].shift(s)
    test_sc_df['shift_{}'.format(s)] = test_sc_df['월별합계'].shift(s)

train_sc_df.head(13)
```

```python
X_train = train_sc_df.dropna().drop('월별합계', axis=1)
y_train = train_sc_df.dropna()[['월별합계']]

X_test = test_sc_df.dropna().drop('월별합계', axis=1)
y_test = test_sc_df.dropna()[['월별합계']]

X_train.head()
```

```python
y_train.head()
```

```python
print(type(X_train))
X_train = X_train.values
print(type(X_train))
X_test= X_test.values

y_train = y_train.values
y_test = y_test.values
print(X_train.shape)
#print(X_train)
print(y_train.shape)
#print(y_train)
```

```python
X_train_t = X_train.reshape(X_train.shape[0], 13, 1)
X_test_t = X_test.reshape(X_test.shape[0], 13, 1)

print("최종 DATA")
print(X_train_t.shape)
print(X_train_t)
print(y_train)
```

```python
from keras.layers import LSTM 
from keras.models import Sequential 
from keras.layers import Dense 
import keras.backend as K 
from keras.callbacks import EarlyStopping

K.clear_session()
    
model = Sequential() # Sequeatial Model 
model.add(LSTM(20, input_shape=(13, 1))) # (timestep, feature) 
model.add(Dense(1)) # output = 1 
model.compile(loss='mean_squared_error', optimizer='adam') 
model.summary()
```

```python
early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)

model.fit(X_train_t, y_train, epochs=100,
          batch_size=30, verbose=1, callbacks=[early_stop])
```

```python
score = model.evaluate(X_test_t,y_test, batch_size=10)
```
