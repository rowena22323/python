```python
from numpy import *
import pandas as pd
import matplotlib.pyplot as plt
import math
from pylab import *
```

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("sin graph")
plt.show()
```

```python
x1 = np.arange(0, math.pi*2, 0.05)
y1 = np.sin(x1)
fig = plt.figure()
ax = fig.add_axes([0.5,0.5,0.5,0.5]) # 그래프 사분면 사이즈 설정
ax.plot(x1, y1)
ax.legend(labels=('a'), loc = 'lower left') # 한글자만 나오나?
ax.set_title('Sine Grpaph')
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
plt.show()
```

```python
fig = plt.figure()
x = np.arange(1,5)
a = fig.subplots(2,2)
# type(a) # 그래프에도 인덱스로 접근이 가능
a[0][0].plot(x, x*x, lw=10)
a[0][0].grid(True)
a[0][0].set_title("square")
a[0][1].plot(x, np.sqrt(x))
a[0][1].grid(color='blue', ls='-.', lw=1)
a[0][1].set_title("square root")
a[1][0].plot(x, np.exp(x))
a[1][1].plot(x, np.log10(x))
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
a5 = fig.add_axes([0.1,0.1,0.8,0.8])
a5.plot(x5, np.exp(x5))
a5.set_ylabel("set")
a6 = a5.twinx() # 보조축
a6.plot(x5, np.log(x5), 'r')
a6.set_ylabel('log')
fig.legend(labels=('exp','log'), loc='upper left')
```

```python
# bar plot
fig = plt.figure(figsize=(9,4))
label = ['java','javascript','python','html5','css3','sql', 'c++','c']
students = [29, 15, 20, 3, 2, 10, 15, 20]
students2 = [25, 15, 10, 9, 10, 15, 1, 15]

x = np.arange(len(label))
a=plt.bar(x-0.0, students, color='g', width=0.2)
b=plt.bar(x+0.2, students2, color='r', width=0.2)

plt.show(a,b)
```

```python
data = [[30,25,50,20],[40,23,51,17],[35,22,45,19]]
x6 = np.arange(4)
fig = plt.figure()
ax6 = fig.add_axes([0,0,1,1])
ax6.bar(x6+0.00, data[0], color='b', width=0.25)
ax6.bar(x6+0.25, data[1], color='g', width=0.25)
ax6.bar(x6+0.50, data[2], color='r', width=0.25)

fig.legend(labels=('developer','dba','se'),loc='best')
plt.show()
```

```python
# 누적그래프
n=5
menMeans=(20, 35, 30, 35, 27)
womenMeans=(25, 32, 34, 20, 25)
ind = np.arange(n)
plt.bar(ind, menMeans, width=0.35, color='lightblue')
plt.bar(ind, womenMeans, width=0.35, bottom=menMeans, color='pink')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend(labels=['Men',"Women"])
plt.show()
```
